```python
import pandas as pd
import numpy as np
import copy
school_df = pd.read_csv("schools_complete.csv")
student_df = pd.read_csv("students_complete.csv")
student_df.head()
```

```python
school_sum_df = pd.DataFrame()

school_sum_df["Total Schools"] = np.array([len(school_df["School ID"])])
school_sum_df["Total Students"] =  np.array([school_df["size"].sum()])
school_sum_df["Average Math Score"] = np.array([student_df["math_score"].mean()])
school_sum_df["Average Reading Score"] = np.array([student_df["reading_score"].mean()])
num_passing_math = 0
for x in student_df["math_score"]:
    
    if x >= 60:
        num_passing_math += 1
percent_passing_math = (num_passing_math/len(student_df["math_score"]))*100
school_sum_df["% Passing Math"] = percent_passing_math
num_passing_reading = 0


percent_passing_reading = (num_passing_reading/len(student_df["reading_score"]))*100
school_sum_df["% Passing Reading"] = percent_passing_reading
school_sum_df["% Overall Passing Rate"]  = (school_sum_df["% Passing Reading"]+school_sum_df["% Passing Math"])/2       
school_sum_df.set_index("Total Students")
school_sum_df


```

```python
schools_indexed = pd.DataFrame()
schools_indexed = school_df
schools_indexed["Per Student Budget"] = schools_indexed["budget"]/schools_indexed["size"]
schools_indexed = schools_indexed.sort_values("name")
schools_indexed["Average Math Score"] = np.array(student_df.groupby("school").math_score.mean())
schools_indexed["Average Reading Score"] = np.array(student_df.groupby("school").reading_score.mean())
passing_averages_math = []
for x in schools_indexed["name"]:
    math_scores_ar = np.array(student_df[student_df.school == x].math_score)
    passing_math = 0
    total = 0
    
    for y in math_scores_ar:
        if y >=70:
            passing_math +=1
            total += 1
        else:
            total += 1
    passing_percent = (passing_math/total) * 100
    passing_averages_math.append(passing_percent)
np.array(passing_averages_math)
schools_indexed["% Passing Math"] = passing_averages_math
passing_averages_reading = []
for x in schools_indexed["name"]:
    reading_scores_ar = np.array(student_df[student_df.school == x].reading_score)
    passing_reading = 0
    total = 0
    
    for y in reading_scores_ar:
        if y >=70:
            passing_reading +=1
            total += 1
        else:
            total += 1
    passing_percent = (passing_reading/total) * 100
    passing_averages_reading.append(passing_percent)
np.array(passing_averages_reading)
schools_indexed["% Passing Reading"] = passing_averages_reading
schools_indexed["% Overall Passing"] = (schools_indexed["% Passing Reading"] + schools_indexed["% Passing Math"])/2
schools_ar = np.array(schools_indexed["name"])
schools_indexed = schools_indexed.set_index("name")
schools_indexed


        
```
